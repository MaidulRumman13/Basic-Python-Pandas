### EDA in Pandas

```python
import pandas as pd
import seaborn as sns
import matplotlib as plt
```

```python
df = pd.read_csv(r"I:\Works\Data Analyst\Basic Python\Pandas\world_population.csv")
df
```

```python
pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

```python
df.info()
```

```python
df.describe()
```

```python
# Count of NULL Values
df.isnull().sum()
```

```python
# Count of Unique Values
df.nunique()
```

```python
# Sorting On Values
df.sort_values(by="2022 Population",ascending=False).head(10)
```

```python
# Correlation Between Columns
df.corr(numeric_only=True)
```

```python
# Heatmap using Seaborn
sns.heatmap(df.corr(numeric_only=True), annot = True)

plt.rcParams['figure.figsize'] = (10,10)

plt.pyplot.show()
```

```python
# Exploring what is Oceania....
df[df['Continent'].str.contains('Oceania')]
```

```python
# Grouping Data
df.groupby('Continent').mean(numeric_only=True).sort_values(by='2022 Population', ascending=False)
```

```python
# Spacified the columns for Visualizing Grouped Data
df2 = df.groupby('Continent')[['1970 Population',
       '1980 Population', '1990 Population', '2000 Population',
       '2010 Population', '2015 Population', '2020 Population',
       '2022 Population']].mean().sort_values(by="2022 Population",ascending=False)

df2
```

```python
# Changing the Columns into Rows & Rows into Columns
df3 = df2.transpose()
df3
```

```python
# Visualizing Data
df3.plot()
```

```python
# Boxplot For Outliers
df.boxplot(figsize = (20,10))
```
